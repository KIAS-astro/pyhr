## Check distribution of positions and velocity

```python
import pyhr as ph

def plot_hist_xv(d, kind, fig=None, axes=None):
    if fig is None and axes is None:
        fig, axes = plt.subplots(2, 3, figsize=(12, 8), constrained_layout=True)
    for i, ax in enumerate(range(3), 1):
        plt.sca(axes[0, i-1])
        plt.hist(d[f'x{i}'], bins=np.linspace(np.min(d[f'x{i}']), np.max(d[f'x{i}']), 50), alpha=0.3)
        plt.xlabel(f'x{i}')
        plt.ylabel(f'# of {kind} ptcls')
        plt.sca(axes[1, i-1])
        plt.hist(d[f'v{i}'], bins=np.linspace(np.min(d[f'v{i}']), np.max(d[f'v{i}']), 50), alpha=0.3)
        plt.xlabel(f'v{i}')
        plt.ylabel(f'# of {kind} ptcls')

    return fig, axes

num = 49
rf = ph.ReadFoF(verbose=False)
df, dfs = rf.get_info(num=num, force_override=False)
hid = 1458183
dat, df_this, indices, df_sub = rf.get_halo(hid)
dd, ds, dg, db = dat[hid]['dm'], dat[hid]['star'], dat[hid]['gas'], dat[hid]['bh']

fig, axes = plot_hist_xv(dd, 'dm')
fig, axes = plot_hist_xv(dg, 'gas', fig, axes)
if len(ds) > 0:
    fig, axes = plot_hist_xv(ds, 'star', fig, axes)
# if len(db) > 0:
#     fig, axes = plot_hist_xv(dg, 'bh', fig, axes)
```
